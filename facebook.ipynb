```python
# Importing Libraries
import pandas as pd
from facebook_scraper import (
    facebook_scraper,get_post_comment
)
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.service import Service
import time
```

```python
#page_list = ["MTCNamibia","TelecomNamibia","paratusnamibia"]
page_list = ["TelecomNamibia"]

proxy_port = 10001
post_counts = 6
browser = 'firefox'
timeout = 600
headless = False

fb_json_data = facebook_scraper(page_list,proxy_port,post_counts,browser,timeout,headless)
```

```python
# Converting  Scrapped results to JSON
df = pd.read_json(fb_json_data)

# Transposing scraped results
df = df.T

# Resetind dataframe index
df.reset_index()

# Renaming columns
mapping = {"Index": 'post_id'} 
post_df = df.rename(columns=mapping) 
```

```python
# Getting facebook comments
comments = get_post_comment(post_df['post_url'].to_list())
```
