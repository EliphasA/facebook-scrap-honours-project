```python
# Importing Libraries
import pandas as pd
import numpy as np
from facebook_scraper import (
    facebook_scraper,get_post_comment
)
```

```python
#page_list = ["MTCNamibia","TelecomNamibia","paratusnamibia"]
telecome_list = ["TelecomNamibia"]
mtc_list = ["MTCNamibia"]
paratus_list = ["paratusnamibia"]

proxy_port = 10001
post_counts = 6000
browser = 'firefox'
timeout = 600
headless = False

fb_tel_json_data = facebook_scraper(telecome_list,proxy_port,post_counts,browser,timeout,headless)
fb_mtc_json_data = facebook_scraper(mtc_list,proxy_port,post_counts,browser,timeout,headless)
fb_paratus_json_data = facebook_scraper(paratus_list,proxy_port,post_counts,browser,timeout,headless)
```

```python
# Converting  Scrapped results to JSON
df_tel = pd.read_json(fb_tel_json_data)
df_mtc = pd.read_json(fb_mtc_json_data)
df_prt = pd.read_json(fb_paratus_json_data)

# Transposing scraped results
df_tel = df_tel.T
df_mtc = df_mtc.T
df_prt = df_prt.T

# Resetind dataframe index
df_tel.reset_index()
df_mtc.reset_index()
df_prt.reset_index()

# Renaming columns
mapping = {"Index": 'post_id'} 
post_tel_df = df_tel.rename(columns=mapping) 

post_mtc_df = df_mtc.rename(columns=mapping) 

post_prt_df = df_prt.rename(columns=mapping) 
```

```python
# Getting facebook comments
comments_tel = get_post_comment(post_tel_df['post_url'].to_list())
comments_mtc = get_post_comment(post_mtc_df['post_url'].to_list())
comments_prt = get_post_comment(post_prt_df['post_url'].to_list())
```
